```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
purchases = pd.read_csv("D:/IDA/Market_Basket_Optimisation.csv")
import warnings
warnings.filterwarnings("ignore")
#1
def apriori_func(df):
    arr=[]
    for index, row in df.iterrows():
        non_zero_values = row.dropna().tolist()
        arr.append(non_zero_values)

    te = TransactionEncoder()
    te_ary = te.fit(arr).transform(arr)
    df_in_new_format = pd.DataFrame(te_ary, columns=te.columns_)
    frequent_itemsets = apriori(df_in_new_format, min_support=0.05, use_colnames=True)

    #2
    rules = association_rules(frequent_itemsets, metric="conviction", min_threshold=0.1)
    return frequent_itemsets, rules

frequent_itemsets, rules = apriori_func(purchases)
print(frequent_itemsets, rules, sep="\n\n", end="\n\n\n")
#3
#В полученных правилах мы видим, что вместе часто покупаются следующие товары:
#    - (spaghetti)	(mineral water)
#    - (chocolate)	(mineral water)	
#В сравнении с оастальными правилами у них высокие показатели метрик: support и confidence
#Эта информация может быть полезной для маркетинговых и продажных стратегий, 
#таких как размещение связанных товаров рядом в магазинах или предложение скидок на сочетания этих товаров, 
#чтобы стимулировать их совместную покупку.

#5

movies_df=pd.read_csv("D:/IDA/movies.csv")
ratings_df=pd.read_csv("D:/IDA/ratings.csv")
#ratings_df=ratings_df.rename(columns={"movieId" : "id"})
#
# ratings_df["id"]=ratings_df["id"].astype("object")
ratings_df=ratings_df.merge(movies_df[["movieId", "title"]], on="movieId", how="left")
# movies_df

data_encoded = ratings_df.groupby(['userId', 'title'])['rating'].sum().unstack().reset_index().fillna(0).set_index('userId')

data_encoded
#frequent_itemsets = apriori(data_encoded, min_support=0.1, use_colnames=True)
df_binary = data_encoded.applymap(lambda x: 1 if x > 3 else 0)

frequent_itemsets = apriori(df_binary, min_support=0.25, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(frequent_itemsets)
rules
```
